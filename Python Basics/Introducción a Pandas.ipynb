```
# Ordenar tablas

pokemon.sort_values('Defense', ascending=False).head()
```

```
ghost_pks = pokemon.loc[pokemon['Type1'] == 'Ghost'].sort_values(['Generation', 'Speed'], ascending=False)

ghost_pks = ghost_pks.reset_index(drop=True)

#ghost_pks.drop(columns=['index'])

ghost_pks
```

```
pokemon.loc[~pokemon['Name'].str.contains('Mega')]
```

```
# Hacer consultas de cadenas de texto usando expresiones regulares
import re 

pokemon.loc[pokemon['Name'].str.contains('f.a+', regex=True, flags = re.I)]
```

# Editar columnas

```
pokemon["Attack-Defense Ratio"] = pokemon["Attack"]/pokemon["Defense"]
```

```
pokemon
```

```
pokemon.loc[pokemon['Type 1'] == 'Poison', 'Type 1'] = 'Venom'
pokemon.head(100)
```

```
pokemon.loc[pokemon['Sp. Atk'] >= 130, 'Legendary'] = True
pokemon.head(50)
```

```
pokemon.loc[pokemon['Type 1'] == 'Water', ['HP', 'Attack']] = [0, -1]
pokemon.head(100)
```

```
pokemon['Speed'].apply(lambda x : x/2).head(100)
#pokemon['Speed'].head(10)
```

```
pokemon.loc[pokemon['Name'].str.contains('Mega')]['Name'].apply(lambda x : 'Mega ' + x.split(' ')[1] if x != 'Meganium' else x)
```

#Agregar columnas

```
pokemon['Total Defense'] = pokemon['Defense'] + pokemon['Sp. Def']
#pokemon = pokemon.drop(columns='Total Defense')
pokemon.head(10)
```

```
# Reajuste de columnas
cols = list(pokemon.columns)

pokemon = pokemon[cols[0:9] + [cols[-1]] + cols[9:12]]
pokemon.head()
```

```
pokemon['Total total'] = pokemon.iloc[:, 4:11].sum(axis = 1)
pokemon.head()
```

# Estadísticas de agregado (groupby)

```
pokemon.groupby(['Type 1', 'Type 2']).mean().sort_values('Defense', ascending=False)
```

```
pokemon.groupby(['Type 1', 'Type 2']).count()['#']
```

```
type_stats = pokemon.groupby(['Type 1']).sum()
```
