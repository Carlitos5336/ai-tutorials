```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

# Un gráfico básico

```
x_values = [1, 2, 3, 4]
y_values = [2, 4, 6, 8]

plt.figure(figsize=(10, 5), dpi=300)

plt.plot(x_values, y_values, label='2x', color='#0c8cb3', marker='s', markersize=5, markeredgecolor='red', linestyle='-')
# Sintaxis ==> [color][marker][linestyle]
#plt.plot(x_values, y_values, 'b*:')

plt.grid()
plt.title("Funciones matemáticas", fontdict={'fontsize':20, 'family':'serif'})
plt.xlabel("Eje de las abcisas")
plt.ylabel("Eje de las ordenadas")
# un linea desde el punto (1,3) hasta (2,4)

plt.xticks([1, 2, 3, 4])
plt.yticks(y_values)

plt.legend()

plt.savefig("funciones.png", dpi=300)

plt.show()
```

```
x = np.arange(0, 9, 0.1)
dx = 2 * x
ex = np.power(np.e, x)
sinx = np.sin(x)

plt.plot(x, sinx)

print(x)

plt.show()
```

```
step = 0.1

x = np.arange(0, 4, step)
y_1 = 2 * x
y_2 = x ** 2

lim = int(2/step)

plt.plot(x[:lim + 1], y_1[:lim + 1], label='2x')
plt.plot(x[:lim + 1], y_2[:lim + 1], label='x^2')

plt.plot(x[lim:], y_1[lim:], 'r--')
plt.plot(x[lim:], y_2[lim:], 'r:')

plt.grid()
plt.title("Funciones matemáticas", fontdict={'fontsize':20, 'family':'serif'})
plt.xlabel("Eje X")
plt.ylabel("Eje Y")

plt.xticks([1, 2, 3, 4])
plt.yticks([1, 4, 9, 16])

plt.legend()
print(x)

plt.show()
```

#Pie Charts

```
left = fifa.loc[fifa['Preferred Foot'] == 'Left'].count()[0]
right = fifa.loc[fifa['Preferred Foot'] == 'Right'].count()[0]

plt.style.use('fivethirtyeight')

plt.pie([left, right], labels=['Izquierdo', 'Derecho'], autopct="%.2f %%")
plt.show()
```

```
def separador(s): # 5'7
  try:
    if s == float('nan'):
      return s
    sep = str(s).split("'") #["5", "7"]
    return int(sep[0]) + int(sep[1])/12
  except:
    #print(s)
    pass

alturas = fifa['Height'].apply(separador).head(100)

pequeños = alturas.loc[alturas < 5.6].count()
medianos = alturas.loc[(alturas >= 5.6) & (alturas < 6.0)].count()
altos = alturas.loc[(alturas >= 6.0) & (alturas < 6.3)].count()
muy_altos = alturas.loc[alturas >= 6.3].count()

plt.figure(figsize=(7, 7))

explode=[0.5, 0, 0, 0.3]

plt.pie([pequeños, medianos, altos, muy_altos], pctdistance=.8 , explode=explode, autopct="%.2f %%", labels=['Menor a 5.6', '5.6-6.0', '6.0-6.3', 'Mayor 6.3'])
plt.title("Cantidad de jugadores de FIFA por altura")

plt.show()
```

#Box and Whiskers

```
barcelona = fifa.loc[fifa['Club'] == 'FC Barcelona']['Overall']
madrid = fifa.loc[fifa['Club'] == 'Real Madrid']['Overall']
juventus = fifa.loc[fifa['Club'] == 'Juventus']['Overall']
manchester_c = fifa.loc[fifa['Club'] == 'Manchester City']['Overall']
chelsea = fifa.loc[fifa['Club'] == 'Chelsea']['Overall']
manchester_u = fifa.loc[fifa['Club'] == 'Manchester United']['Overall']

plt.xticks(rotation=90)

plt.title("Comparación de equipos de FIFA")
plt.boxplot([barcelona, madrid, juventus, chelsea, manchester_c, manchester_u], labels=["Barcelona", "Real Madrid", "Juventus", "Chelsea", "Manchester City", "Manchester United"])
plt.show()
```
